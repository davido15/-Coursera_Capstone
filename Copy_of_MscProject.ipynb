```python
filename = 'data.csv'
df = pd.read_csv(filename)
df.describe()
```

Feature selection

```python
#Preview the floor column
df.columns.get_loc("FLOOR")
#input data from WPA001 to WPA520
X= df.iloc[:, :519]
#value to be predicted
y= df.loc[:, 'FLOOR']
```

Split the data into 75 percent  training and 25 % test

```python
#split data testing and training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

```

```python
y_test.describe()
```

```python
X_test.describe()
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# Fit only on X_train
scaler.fit(X_train)

# Scale both X_train and X_test
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
Xvalidate = scaler.transform(Xvalidate)
```

Classification based on KNN

```python
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.neighbors import KNeighborsRegressor
knnalgo = KNeighborsRegressor(n_neighbors=1)
knnalgo.fit(X_train, y_train)
```

```python
# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

```python
y_pred = knnalgo.predict(Xvalidate)
print(y_pred)
```

```python
y_pred = knnalgo.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

knn_mae = mean_absolute_error(y_test, y_pred)
knn_mse = mean_squared_error(y_test, y_pred)
knn_rmse = mean_squared_error(y_test, y_pred, squared=False)

print(f'mae: {knn_mae}')
print(f'mse: {knn_mse}')
print(f'rmse: {knn_rmse}')
```

```python
# Model Accuracy: how often is the classifier correct?
knn_accuracy=metrics.accuracy_score(y_test, y_pred)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Classification based on SVM

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import multilabel_confusion_matrix, ConfusionMatrixDisplay


#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

svm_mae = mean_absolute_error(y_test, y_pred)
svm_mse = mean_squared_error(y_test, y_pred)
svm_rmse = mean_squared_error(y_test, y_pred, squared=False)

print(f'mae: {svm_mae}')
print(f'mse: {svm_mse}')
print(f'rmse: {svm_rmse}')
```

```python


# Model Accuracy: how often is the classifier correct?

svm_accuracy=metrics.accuracy_score(y_test, y_pred)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
